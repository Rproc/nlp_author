```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, KFold
from sklearn import preprocessing
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from statistics import mean 

from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
import warnings

warnings.simplefilter('ignore')
```

## Bicicle Sharing.
### The idea is to how many bicicles will be allocated for use in any day.

#### The main goal is to know the number of users of the system in determined days, for that there is a series of features such as temperature, real temperature, climate, day of the week, if it's a holiday or weekend or not, etc.


```python
bk_sharing = pd.read_csv('day.csv')
```

```python
bk_sharing
```

```python
bk_sharing.columns
```

```python
X = bk_sharing[['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday',
       'weathersit', 'temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered']]

aux = X
```

```python
Y = bk_sharing[['cnt']]
```

```python
season_binarized = pd.DataFrame(preprocessing.LabelBinarizer().fit_transform(X.season))
month_binarized = pd.DataFrame(preprocessing.LabelBinarizer().fit_transform(X.mnth))
weekday_binarized = pd.DataFrame(preprocessing.LabelBinarizer().fit_transform(X.weekday))
weathersit_binarized = pd.DataFrame(preprocessing.LabelBinarizer().fit_transform(X.weathersit))

X_nbin = X[['yr', 'holiday', 'workingday', 'temp', 'atemp', 'hum', 'windspeed',
       'casual', 'registered']]

X_bin = pd.concat([season_binarized, month_binarized, weekday_binarized, weathersit_binarized], axis=1)

X = pd.concat([X_nbin,X_bin],axis=1)
```

```python
model = svm.SVR()
kfold = KFold(n_splits=10, shuffle=False, random_state=42)

error = []
error_r = []
score_list = []
# real_target = []
pred_target = []
# pred = np.asarray(pred_target)

for i, (train, test) in enumerate(kfold.split(X, Y)):
    model.fit(X.iloc[train,:], Y.iloc[train,:])
    predicted = model.predict(X.iloc[test,:])
    m = mean_absolute_error(predicted,Y.iloc[test,:])
    r = mean_squared_error(predicted,Y.iloc[test,:])
    score = model.score(X.iloc[test,:], Y.iloc[test,:])
    
    pred_target.append(predicted)
    error.append(m)
    error_r.append(r)
    score_list.append(score)

    
```

```python
mae = mean(error)
rmse = mean(error_r)
score = mean(score_list)

npArray = np.array([ elem for singleList in pred_target for elem in singleList])
y2 = pd.DataFrame(npArray)
```

```python
print(mae)
print(rmse)
print(score)

print('Y (expected) --- Mean Value:', np.mean(Y), 'Std.: ', np.std(Y))
print('y2 (regression) --- Mean Value:', np.mean(y2), 'Std.: ', np.std(y2))
```
